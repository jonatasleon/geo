This is my first [Jupyter Notebook](https://jupyter-notebook.readthedocs.io/en/stable/notebook.html) handling spatial data. It uses GeoPandas to [plot a world map](http://geopandas.org/mapping.html#mapping-tools).

```python
import geopandas as gpd
```

- List available datasets:

```python
gpd.datasets.available
```

- Load a dataset from [Natural Earth](https://www.naturalearthdata.com/):

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

- Check available columns:

```python
world.keys()
```

- Calculate the GDP per capita:

```python
world = world[(world.pop_est > 0) & (world.name != 'Antarctica')]
world['gdp_per_capita'] = world.gdp_md_est / world.pop_est
```

- Plot a [cloropleth map](http://geopandas.org/mapping.html#choropleth-maps) whose [colormap](https://matplotlib.org/users/colormaps.html) varies according to the GDP per capita:

```python
world.plot(column = 'gdp_per_capita', cmap = 'Greens', scheme = 'Quantiles')
```
